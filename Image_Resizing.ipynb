```python
#Upload image
from google.colab import files
uploaded = files.upload()
```

```python
# Get the name of the uploaded image
image_name = list(uploaded.keys())[0]
# Print the image name
print(image_name)

image_data = uploaded[image_name]
```

```python
resize_scale = 2
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Decode the image data
image = cv2.imdecode(np.frombuffer(image_data, np.uint8), cv2.IMREAD_UNCHANGED)

# Get original image dimensions
height, width, channels = image.shape
title = f"Original Image ({width}x{height})"

# Resize to specific dimensions
resized_image = cv2.resize(image, (int(width*resize_scale), int(height*resize_scale)))

# Get reshaped image dimensions
resized_height, resized_width, channels = resized_image.shape
resized_title = f"Original Image ({resized_width}x{resized_height})"

# Display
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert to RGB for correct display
plt.title(title)
plt.show()
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))  # Convert to RGB for correct display
plt.title(resized_title)
plt.show()
```
